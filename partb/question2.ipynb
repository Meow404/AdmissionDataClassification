# PartB: Regression Problem

## 0. Read Data

```python
import pandas as pd
data = pd.read_csv("./data/admission_predict.csv")
data = pd.DataFrame(data, columns = data.columns.delete(0))
data
```

# Correlation Matrix

```python
corr = data.corr()
corr
```

## Correlated Features

The pair of features are most correlated to each other

- `CGPA` ~ `Chance of Admit`

- `GRE Score` ~ `TOEFL Score`

- `GRE Score` ~ `CGPA`

The feature has highest correlation coefficient with the `Chance of Admit` is 

- `CGPA`

Descending order:

- `CGPA` (5)
- `GRE Score` (0)
- `TOEFL Score` (1)
- `University Rating` (2)
- `LOR` (4)
- `SOP` (3)
- `Research` (6)

```python
import matplotlib.pyplot as plt
f = plt.figure(figsize=(12.8, 10.8))
plt.matshow(corr, fignum=f.number)
plt.xticks(range(data.shape[1]), data.columns, fontsize=11, rotation=45)
plt.yticks(range(data.shape[1]), data.columns, fontsize=11)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=12);
```
